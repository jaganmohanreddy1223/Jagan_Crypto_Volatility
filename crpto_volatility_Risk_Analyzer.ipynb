```python
%%html
from flask import Flask, jsonify, render_template
import requests
import pandas as pd
from datetime import datetime

app = Flask(__name__)

def fetch_crypto_data():
    url = "https://api.coingecko.com/api/v3/coins/markets"
    params = {
        "vs_currency": "usd",
        "order": "market_cap_desc",
        "per_page": 5,
        "page": 1,
        "sparkline": False
    }

    response = requests.get(url, params=params)
    data = response.json()

    crypto_list = []
    for coin in data:
        crypto_list.append({
            "name": coin["name"],
            "symbol": coin["symbol"].upper(),
            "price": coin["current_price"],
            "change": coin["price_change_percentage_24h"],
            "volume": coin["total_volume"]
        })

    df = pd.DataFrame(crypto_list)
    df.fillna(0, inplace=True)
    df["time"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    df.to_csv("crypto_data.csv", index=False)

    return crypto_list

@app.route("/")
def home():
    return render_template("index.html")

@app.route("/data")
def data():
    crypto_data = fetch_crypto_data()
    return jsonify(crypto_data)

if __name__ == "__main__":
    app.run(debug=True)


<!DOCTYPE html>
<html>
<head>
    <title>Milestone 1 - Data Acquisition</title>
    <meta charset="UTF-8">
    <style>
        body{
            background:linear-gradient(135deg,#050b24,#0b1c44);
            color:white;
            font-family:Arial;
            margin:0;
            padding:30px;
        }
        .container{
            display:grid;
            grid-template-columns:1fr 1.4fr;
            gap:30px;
            max-width:1300px;
            margin:auto;
        }
        .card{
            background:rgba(255,255,255,0.06);
            border-radius:16px;
            padding:25px;
        }
        h1,h2{margin-top:0}
        ul{list-style:none;padding:0}
        li{margin:12px 0}
        button{
            background:#35d3ff;
            border:none;
            padding:8px 14px;
            border-radius:8px;
            font-weight:600;
            cursor:pointer;
        }
        table{width:100%;border-collapse:collapse;margin-top:15px}
        th,td{padding:10px;text-align:left}
        th{color:#35d3ff}
        tr{border-bottom:1px solid rgba(255,255,255,0.1)}
    </style>
</head>
<body>

<div class="container">

    <div class="card">
        <h1>Milestone 1: Data Acquisition</h1>

        <h3>Requirements</h3>
        <ul>
            <li>Python Environment Setup</li>
            <li>API Integration (CoinGecko)</li>
            <li>CSV Data Storage</li>
            <li>Missing Value Handling</li>
        </ul>

        <h3>Outputs</h3>
        <ul>
            <li>Live Crypto Prices</li>
            <li>Verified API Connectivity</li>
            <li>Auto CSV Storage</li>
        </ul>
    </div>

    <div class="card">
        <div style="display:flex;justify-content:space-between;align-items:center">
            <h2>Crypto Data Fetcher ● Live</h2>
            <button onclick="loadData()">Refresh</button>
        </div>

        <p id="time"></p>

        <table>
            <thead>
                <tr>
                    <th>Crypto</th>
                    <th>Price ($)</th>
                    <th>24h %</th>
                    <th>Volume</th>
                </tr>
            </thead>
            <tbody id="tableBody"></tbody>
        </table>
    </div>

</div>

<script>
function loadData(){
    fetch("/data")
    .then(res => res.json())
    .then(data => {
        let rows = "";
        data.forEach(c=>{
            rows += `<tr>
                <td>${c.name} (${c.symbol})</td>
                <td>$${c.price}</td>
                <td>${c.change.toFixed(2)}%</td>
                <td>${c.volume}</td>
            </tr>`;
        })
        document.getElementById("tableBody").innerHTML = rows;
        document.getElementById("time").innerText = "Last Updated: " + new Date().toLocaleString();
    });
}
loadData();
</script>

</body>
</html>
```
